```python
Q1. Explain GET and POST methods.
GET Method:

Purpose: The GET method is used to request data from a specified resource. It retrieves information from the server based on the URL provided.
Characteristics:
Parameters are sent in the URL (query string).
It is considered safe and idempotent, meaning it does not modify data on the server.
Data sent via GET is visible in the URL and has length limitations.
POST Method:

Purpose: The POST method is used to send data to the server to create or update a resource. It submits data to be processed to the specified resource.
Characteristics:
Parameters are sent in the body of the request, not in the URL.
It is not idempotent, meaning it can modify or create resources on the server.
Data sent via POST is not visible in the URL and has no size limitations.
Q2. Why is request used in Flask?
request in Flask:

The request object in Flask is used to access data from incoming HTTP requests. It allows you to retrieve data sent by the client, such as form data, query parameters, and file uploads.
Common Uses:
Access form data using request.form.
Retrieve query parameters using request.args.
Handle file uploads with request.files.
Obtain request headers and cookies.
Example:

from flask import Flask, request

app = Flask(__name__)

@app.route('/form', methods=['POST'])
def form():
    name = request.form.get('name')
    return f'Hello, {name}'

if __name__ == '__main__':
    app.run(debug=True)
In this example, request.form.get('name') retrieves the value of the name field from a submitted form.

Q3. Why is redirect() used in Flask?
redirect() in Flask:

The redirect() function is used to redirect the user to a different route or URL. It generates an HTTP redirect response to the specified location.
Common Uses:
Redirect users after form submission to avoid resubmission.
Redirect to a different page based on conditions or permissions.
Redirect to an external URL.
Example:

from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/redirect')
def redirect_to_home():
    return redirect(url_for('home'))

if __name__ == '__main__':
    app.run(debug=True)
In this example, redirect(url_for('home')) sends a user to the / route.

Q4. What are templates in Flask? Why is the render_template() function used?
Templates in Flask:

Templates in Flask are HTML files used to define the structure and layout of web pages. They allow dynamic content generation by embedding Python code within HTML.
Flask uses the Jinja2 templating engine to render templates, allowing for variable substitution, control structures, and reusable components.
render_template() Function:

The render_template() function is used to render a template with context variables. It generates the final HTML by combining the template with data passed from the view function.
Example:

Template File (templates/index.html):

html
Copy code
<!DOCTYPE html>
<html>
<head>
    <title>{{ title }}</title>
</head>
<body>
    <h1>{{ heading }}</h1>
    <p>{{ message }}</p>
</body>
</html>
Flask Application:

from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html', title='Home Page', heading='Welcome!', message='Hello, world!')

if __name__ == '__main__':
    app.run(debug=True)
In this example, render_template('index.html', title='Home Page', heading='Welcome!', message='Hello, world!') renders the index.html template with the specified context variables.

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.
Flask API Example:

Code:


from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/api/greet', methods=['GET'])
def greet():
    name = request.args.get('name', 'Guest')
    return jsonify({'message': f'Hello, {name}!'})

if __name__ == '__main__':
    app.run(debug=True)
Testing with Postman:

```
